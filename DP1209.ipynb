```
#鐵條切割
def CutRod(p,n):
    r = [0]*(n+1)
    for i in range(1, n+1):
        if n == 0:
            return 0
        q =0
        for j in range(1, i+1):
            q = max(q, p[j]+r[i-j])
            r[i] = q
    return r[n]
p=[0,1,5,8,9,10,17,17,20,24,30]
n = int(input("輸入切割長度:"))
print(CutRod(p,n))
```
